## import the library

```python
import mysql.connector as connector
import pandas as pd
```

# create a connection to database 

```python
try: 
    conn = connector.connect(host="localhost", db="mysql" ,user="root" ,password="Mysql123")
except connector.Error as e: 
    print("Error: Could not make connection to the Mysql database")
    print(e)
```

# Use the connection to get a cursor that can be used to execute queries.

```python
try: 
    cur = conn.cursor()
except connector.Error as e: 
    print("Error: Could not get curser to the Database")
    print(e)
```

# Set automatic commit to be true so that each action is committed without having to call conn.commit() after each command.

```python
conn.autocommit=True
```

## Create a database

```python
try: 
    cur.execute("create database wealthinfo")
except connector.Error as e:
    print(e)
```

## Closing the connection to default mysql database and creating connection to new database

```python
try: 
    conn.close()
except connector.Error as e:
    print(e)
    
try: 
    conn = connector.connect(host="localhost", db="wealthinfo" ,user="root" ,password="Mysql123")
except connector.Error as e: 
    print("Error: Could not make connection to the mysql database")
    print(e)
    
try: 
    cur = conn.cursor()
except connector.Error as e: 
    print("Error: Could not get cursor to the Database")
    print(e)
```

```python
accounts_country_table_create=("""CREATE TABLE IF NOT EXISTS accountscountry(
country_code varchar(50),
short_name varchar(50),
table_name varchar(50),
Long_name varchar(100),
currency_unit varchar(50)
)""")
```

```python
cur.execute(accounts_country_table_create)
```

```python
account_data_table_create=("""CREATE TABLE IF NOT EXISTS accountdata(
country_name varchar(50),
country_code varchar(50),
series_name varchar(50),
series_code varchar(50),
year_2000 varchar(50),
year_2010 varchar(50),
year_2018 varchar(50)
)""")
```

```python
cur.execute(account_data_table_create)
```

```python
accounts_series=("""CREATE TABLE IF NOT EXISTS accountseries(
code varchar(50),
topic varchar(50),
indicator_name varchar(500))""")
```

```python
cur.execute(accounts_series)
```

```python
accounts_country_table_insert=("""INSERT INTO accountscountry(
country_code,
short_name,
table_name,
Long_name,
currency_unit)
VALUES(%s,%s,%s,%s,%s)
""")
```

```python
for i,row in AccountsCountry.iterrows():
    #print(list(row))
    cur.execute(accounts_country_table_insert,list(row))
```
